### Neural Network using Keras

````python
import random
import pandas as pd
import numpy  as np

X = np.array([[1,1,0,0],[0,1,1,0],[0,0,0,1]])
y = np.array([[1,0,0,0],[0,0,1,1],[0,1,1,0]])
X_cv = np.array([[0, 0, 0, 1]])
````

````python
# Creating a model
from keras.models import Sequential
from keras.layers import Dense, Activation
import random

model = Sequential()

model.add(Dense(units=4,input_dim=4))
model.add(Activation('sigmoid'))
````

````python
model.summary(70)
````

````python
# Compiling model
from keras.optimizers import SGD
import random

model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.01, momentum=0.9, decay=0.0, nesterov=False), metrics=['mean_squared_error'])
````

````python

hist = model.fit(X, y, epochs=200, verbose=2, validation_split=0.3)
````

````python
print(model.predict(np.array([[1, 1, 0, 0]])))
````

````python
print(model.predict(np.array([[0, 1, 1, 0]])))
````

````python
print(model.predict(np.array([[0, 0, 0, 1]])))
````

````python
import matplotlib.pyplot as plt
mse_test = list(reversed(hist.history['val_mean_squared_error']))

# summarize history for loss
plt.plot(hist.history['mean_squared_error'])
plt.plot(mse_test)
plt.title('model MSE')
plt.ylabel('MSE')
plt.xlabel('hidden neurons')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
````

** Results **
Se evidencia que a medida que el número de neuronas aumenta, en el set de Train tiende a haber un Overfitting, mientras que en el set de test el MSE tiende a disminuir, pero sin llegar a ser cero.
Por otro lado, se realizó la comprobación de los resultados del modelo con los datos de Train y en el último, dió un resultado diferente al modelo por medio de sklearn. Se debe tener en cuenta, que el número de neuronas pudo no ser el adecuado y que hace falta realizar un tuneo al modelo.

## 5.2 Boston House Price Prediction

In the next questions we are going to work using the dataset *Boston*. This dataset measures the influence of socioeconomical factors on the price of several estates of the city of Boston. This dataset has 506 instances, each one characterized by 13 features:

* CRIM - per capita crime rate by town
* ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
* INDUS - proportion of non-retail business acres per town.
* CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
* NOX - nitric oxides concentration (parts per 10 million)
* RM - average number of rooms per dwelling
* AGE - proportion of owner-occupied units built prior to 1940
* DIS - weighted distances to five Boston employment centres
* RAD - index of accessibility to radial highways
* TAX - full-value property-tax rate per 10,000 USD
* PTRATIO - pupil-teacher ratio by town
* B - $1000(Bk - 0.63)^2$ where $Bk$ is the proportion of blacks by town
* LSTAT - % lower status of the population

Output variable:
* MEDV - Median value of owner-occupied homes in 1000's USD

**Note:** In this exercise we are going to predict the price of each estate, which is represented in the `MEDV` variable. It is important to remember that we are always aiming to predict `MEDV`, no matter which explanatory variables we are using. That means, in some cases we will use a subset of the 13 previously mentioned variables, while in other cases we will use all the 13 variables. But in no case we will change the dependent variable $y$.

### Perceptron Algorithm
https://medium.com/technology-invention-and-more/how-to-build-a-multi-layered-neural-network-in-python-53ec3d1d326a
https://machinelearningmastery.com/implement-perceptron-algorithm-scratch-python/

````python
from sklearn.neural_network import MLPRegressor
mlr = MLPRegressor(solver='lbfgs', alpha=1e-5, 
                     hidden_layer_sizes=(128,512), random_state=1)
mlr.fit(X_train, y_train)
mlrpred = mlr.predict(X_test)
````

````python
from sklearn.metrics import mean_squared_error, r2_score

print('MSE:',mean_squared_error(y_test, mlrpred))
print('Variance score: %.2f' %  r2_score(y_test, mlrpred))
````

````python
from sklearn.neural_network import MLPRegressor
mlr_s = MLPRegressor(solver='lbfgs', alpha=1e-5, 
                     hidden_layer_sizes=(128,512), random_state=1)
mlr_s.fit(X_train_s, y_train)
mlrpred_s = mlr_s.predict(X_test_s)
````

````python
from sklearn.metrics import mean_squared_error, r2_score

print('MSE:',mean_squared_error(y_test, mlrpred_s))
print('Variance score: %.2f' %  r2_score(y_test, mlrpred_s))
````

** Which model has better performance? Why? **
El modelo que tiene un mejor performance es el 2-layer MultiLayer Perceptron (128 neurons in the first layer and 512 in the second layer) con los datos escalados. Esto sucede porque al reducir la varianza de los datos y poder eliminar el efecto de las unidades de las variables, permite al algoritmo converger de una mejor manera (gradiente) a aquel punto con un mínimo global. 
El modelo dió como resultado un MSE de 13.889 y un R^2 de 0.85, que es superior al de su par sin datos escalados y que el de los modelos de Regresión Lineal.
De igual manera, es importante tener en cuenta las dos capas de redes neuronales que permiten una predicción ajustada de los datos, sin incurrir en Overfitting.
